```python
# Paso 1: importar librerías
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

# Paso 2: cargar dataset desde sklearn
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# Paso 3: primeras filas
df.head()
```

```python
# Información general del dataset
print("Dimensiones:", df.shape)
print("\nDescripción estadística:")
print(df.describe())

# Balance de la variable objetivo (target: 0 = maligno, 1 = benigno)
print("\nDistribución de la variable objetivo:")
print(df['target'].value_counts())
```

```python
# Histograma de un par de variables
df[['mean radius', 'mean texture']].hist(figsize=(8,4))
plt.suptitle("Distribución de variables básicas")
plt.show()

# Gráfico de barras: distribución de la variable objetivo
df['target'].value_counts().plot(kind='bar')
plt.title("Distribución: 0 = Maligno, 1 = Benigno")
plt.xlabel("Clase")
plt.ylabel("Frecuencia")
plt.show()
```
